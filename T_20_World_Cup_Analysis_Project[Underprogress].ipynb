#### Subset Selection:
We will narrow down our focus to columns related to player statistics and contextual information:

#### Handling Missing Data:
- Evaluate the presence of missing values in the selected columns.
- Decide on appropriate strategies for imputation or removal based on the impact on analysis.

#### Feature Engineering:
- Explore opportunities for creating new features derived from existing ones.
- Consider encoding categorical variables and transforming data as needed.

**Note:** The subset selection and data preparation steps lay the foundation for a focused analysis on key player statistics and contextual information. The subsequent phases will delve into imputation, further analysis, and interpretation of insights.


```python
import pandas as pd

# Select relevant columns
columns_of_interest = ['match_id', 'innings_id', 'runs', 'batsman1_runs', 'batsman2_runs',
                       'bowler1_wkts', 'bowler2_wkts', 'over', 'ball', 'isNoball', 'isWide', 'isBoundary']

# Subset the data
subset_data = data[columns_of_interest]

# Drop rows with missing values in relevant columns
subset_data.dropna(subset=['runs', 'batsman1_runs', 'batsman2_runs', 'bowler1_wkts', 'bowler2_wkts'], inplace=True)

subset_data.head()
```

## 2. Data Analysis and Imputation

#### Imputation Strategies:

```python
# Impute missing values in the 'isNoball' and 'isWide' columns
subset_data['isNoball'].fillna(0, inplace=True)
subset_data['isWide'].fillna(0, inplace=True)

# Interpolate missing values in 'over' and 'ball' columns to maintain the sequence
subset_data['over'].interpolate(method='linear', inplace=True)
subset_data['ball'].interpolate(method='linear', inplace=True)

# Forward-fill missing values in other columns to propagate values from the previous non-missing row
subset_data.fillna(method='ffill', inplace=True)

subset_data.head()
```

We address missing values in the 'isNoball' and 'isWide' columns by filling NaN values with 0, assuming a default of no no-ball or wide.
For 'over' and 'ball' columns, missing values are interpolated linearly to maintain the sequence and ensure a logical progression.
Other missing values in the dataset are forward-filled, propagating values from the previous non-missing row to maintain continuity.


##3. Data Analysis

Now that we've prepared the data and imputed missing values, we can perform in-depth analysis to examine player statistics. We can calculate player-wise statistics such as total runs scored, wickets taken, and over-by-over progress for each match and innings.

```python
# Calculate total runs scored by each batsman
subset_data['batsman1_total_runs'] = subset_data.groupby(['match_id', 'innings_id'])['batsman1_runs'].cumsum()
subset_data['batsman2_total_runs'] = subset_data.groupby(['match_id', 'innings_id'])['batsman2_runs'].cumsum()
subset_data['batsman2_total_runs']
```

```python
# Calculate cumulative wickets taken by each bowler in each match and innings
subset_data['bowler1_cumulative_wkts'] = subset_data.groupby(['match_id', 'innings_id'])['bowler1_wkts'].cumsum()
subset_data['bowler2_cumulative_wkts'] = subset_data.groupby(['match_id', 'innings_id'])['bowler2_wkts'].cumsum()


print("Cumulative wickets taken by bowler_1: ", subset_data['bowler1_cumulative_wkts'] )
print("Cumulative wickets taken by bowler_1: ", subset_data['bowler2_cumulative_wkts'])
```

```python
# Calculate total runs scored in each over
over_progress = subset_data.groupby(['match_id', 'innings_id', 'over'])[['runs']].sum().reset_index()

# Calculate total wickets taken in each over
over_progress['wickets'] = subset_data.groupby(['match_id', 'innings_id', 'over'])[['bowler1_wkts', 'bowler2_wkts']].sum().sum(axis=1).values

print("Total runs scored in each over : ", over_progress)
print("Total wickets taken in each over: ", over_progress['wickets'])
```
